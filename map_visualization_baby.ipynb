```python
df = pd.read_csv('/content/drive/MyDrive/데이터포럼/Toilet_Data/서울교통공사_역사공중화장실정보_20220331.csv', encoding='cp949')
print(df.head(5))
```

```python
w = []
m = []
b = []

for i in range(len(df['place'])):
    tmp = []
    if (df['place'][i] == "여자화장실"):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        tmp.append(df['place'][i])
        w.append(tmp)
    if (df['place'][i] == "남자화장실"):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        tmp.append(df['place'][i])
        m.append(tmp)
    if (df['place'][i] == "남자화장실,여자화장실"):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        tmp.append(df['place'][i])
        b.append(tmp)
    else:
      pass
```

```python
# 기저귀 교환대가 여자화장실 둘 다에 있는 경우

map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(w)):
     folium.Marker([float(w[i][0]),float(w[i][1])], popup='여자화장실', icon=folium.Icon(color='red', icon='info-sign')).add_to(map_osm)

map_osm
```

```python
# 기저귀 교환대가 남자화장실에 있는 경우

map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(m)):
     folium.Marker([float(m[i][0]),float(m[i][1])], popup='남자화장실', icon=folium.Icon(color='blue', icon='info-sign')).add_to(map_osm)

map_osm
```

```python
# 기저귀 교환대가 남자화장실, 여자화장실 둘 다에 있는 경우

map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(b)):
     folium.Marker([float(b[i][0]),float(b[i][1])], popup='남자화장실,여자화장실 둘 다', icon=folium.Icon(color='purple', icon='info-sign')).add_to(map_osm)

map_osm
```
